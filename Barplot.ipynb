```python
from matplotlib import pyplot as plt
x=[1,3,6,7,9]
y=[10,11,12,13,14]
plt.plot(x,y)
plt.title("stock exchange")
plt.xlabel("x-axis")
plt.ylable("y-axis")
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,3,6,7]
x1=[9,12,15]
plt.plot(x,"g",label="line one",linewidth=5)
plt.plot(x1,"r",label="line two",linewidth=5)
plt.title("stock exchange")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(color="k")
plt.show()
```
